```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('09-19-Amazon-Ranking-Analysis.csv')
data['difference'] = data.CorrectedPrice - data.ScrapedIndexPrice
data.sort_values('difference', ascending=False)
```

```python

type = data[['ScrapedIndexVendorType', 'difference']]
type.sort_values('difference', ascending=False)
amazon = type[(type.ScrapedIndexVendorType == 'Amazon') | (type.ScrapedIndexVendorType == 'FBA')]
not_amazon = type[~(type.ScrapedIndexVendorType == 'Amazon') & ~(type.ScrapedIndexVendorType == 'FBA')]
amazon.plot.hist(title='Amazon', range=[0,30], facecolor='grey', align='left')
plt.xlabel('Price')
plt.savefig('amazon.jpg')

```

```python

not_amazon = type[~(type.ScrapedIndexVendorType == 'Amazon') & ~(type.ScrapedIndexVendorType == 'FBA')]
not_amazon.plot.hist(title='Third Party', range=[0,30], align='left')
plt.xlabel('Price')
plt.savefig('thirdparty.jpg')
```
